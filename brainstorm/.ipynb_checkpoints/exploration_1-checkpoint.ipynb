```python
import csv

def generate_insert_statements(csv_file):
    insert_statements = []
    with open(csv_file, 'r', newline='') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Skip header row
        for row in reader:
            values = ', '.join([f"'{value}'" for value in row])
            insert_statement = f"INSERT INTO Climate_Metrics VALUES ({values});"
            insert_statements.append(insert_statement)
    return insert_statements


def write_insert_statements_to_file(insert_statements):
    with open('create_tables.sql', 'w') as file:  # Append mode
        for statement in insert_statements:
            file.write(statement + '\n')
```

```python
# Step 2: Generate INSERT statements
insert_statements = generate_insert_statements("data/climate_toy.csv")

# Step 3: Write INSERT statements to the same file
write_insert_statements_to_file(insert_statements)
```
