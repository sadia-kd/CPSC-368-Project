```python
import pandas as pd
import numpy as np
import altair as alt
```

# 1 : climate

```python
climate = pd.read_csv("data/climate_change_data.csv", parse_dates = True)
climate.shape
```

```python
climate = climate[climate['Date'] >= '2020']
climate.tail()
```

```python
climate.columns
```

```python
climate.dtypes
```

```python
climate['Country'].nunique()
climate[climate['Country'] == 'Afghanistan']
```

```python
alt.Chart(climate).mark_bar().encode(
    x = alt.X('Country').sort('-y'),
    y = alt.Y('Sea Level Rise'),
    #tooltip = alt.Tooltip('Location')
    #color = alt.Color('Temperature')
    
)
```

```python
alt.Chart(climate).mark_point().encode(
    x = 'Precipitation',
    y = 'Temperature',
    #color = 'Country'
)
```

```python
# no trend, all cluttered in one area
alt.Chart(climate).mark_point().encode(
    x = 'CO2 Emissions',
    y = 'Temperature',
    color = 'Sea Level Rise'
).interactive()
```

```python
alt.Chart(climate).mark_point().encode(
    x = 'CO2 Emissions',
    y = 'Sea Level Rise',
    #color = 'Country'
).interactive()
```
